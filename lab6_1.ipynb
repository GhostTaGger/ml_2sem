```python
class Neuron:
    """Искусственный нейрон"""

    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias

    def activate(self, inputs):
        total_input = sum(w * i for w, i in zip(self.weights, inputs)) + self.bias
        return 1 if total_input >= 0 else 0


# Базовые логические элементы
neuron_and = Neuron(weights=[1, 1], bias=-1.5)   # AND: 1 только при (1,1)
neuron_or = Neuron(weights=[1, 1], bias=-0.5)    # OR:  1 при любом 1
neuron_not = Neuron(weights=[-1], bias=0.5)      # NOT: инверсия

def xor_network(x1, x2):
    """Реализация XOR через комбинацию нейронов"""
    # Ветка 1: x1 AND NOT x2
    not_x2 = neuron_not.activate([x2])
    branch1 = neuron_and.activate([x1, not_x2])
    
    # Ветка 2: NOT x1 AND x2
    not_x1 = neuron_not.activate([x1])
    branch2 = neuron_and.activate([not_x1, x2])
    
    # Объединение: branch1 OR branch2
    return neuron_or.activate([branch1, branch2])

# Проверка всех комбинаций
print("XOR через сеть нейронов:")
for x1, x2 in [(0, 0), (0, 1), (1, 0), (1, 1)]:
    output = xor_network(x1, x2)
    print(f'Вход: [{x1}, {x2}] → Выход: {output} (ожидается {x1 ^ x2})')
```

```python
import random

def init_weights(num_weights=2):
    """Инициализирует веса случайными значениями из [0, 1)."""
    return [random.random() for _ in range(num_weights)]

def predict(x, w):
    """Предсказывает результат на основе весов."""
    summator = x[0] * w[0] + x[1] * w[1]
    return 1 if summator >= 1 else 0

def train(x_train, y_train, w, speed):
    """Обучает модель методом коррекции ошибки."""
    for i in range(len(x_train)):
        error = y_train[i] - predict(x_train[i], w)
        if error != 0:
            w[0] = w[0] + error * x_train[i][0] * speed
            w[1] = w[1] + error * x_train[i][1] * speed

def epoch(count, func_train, x_train, y_train, w, speed):
    """Выполняет указанное количество эпох обучения."""
    print("Эволюция весов в процессе обучения:")
    for i in range(count):
        func_train(x_train, y_train, w, speed)
        print(f"Эпоха {i+1}: w = [{w[0]:.4f}, {w[1]:.4f}]")

# Данные для обучения (полная таблица истинности OR)
x_train = [[0, 0], [0, 1], [1, 0], [1, 1]]
y_train = [0, 1, 1, 1]

# Инициализация весов и параметров обучения
random.seed(42)  # Для воспроизводимости
weights = init_weights(2)
learning_rate = 0.5
epochs = 10

# Обучение
epoch(epochs, train, x_train, y_train, weights, learning_rate)

# Финальная проверка на всех комбинациях
print("\nПроверка работы нейрона после обучения:")
for x in x_train:
    output = predict(x, weights)
    expected = x[0] or x[1]
    status = "✓" if output == expected else "✗"
    print(f"Вход: {x} → Выход: {output} (ожидается {expected}) {status}")

# Альтернатива: прямая инициализация оптимальных весов
print("\n--- Альтернативный вариант: прямая инициализация ---")
weights_optimal = [1.0, 1.0]
print(f"Веса: {weights_optimal}")
for x in x_train:
    output = predict(x, weights_optimal)
    expected = x[0] or x[1]
    print(f"Вход: {x} → Выход: {output} (ожидается {expected})")
```
